```python
import numpy as np
import tensorflow as tf
```

```python
np.set_printoptions(linewidth=100, precision=4)
sess = tf.InteractiveSession()
```

```python
num_batches = 3
num_detections = 28
num_classes = 4
```

```python
a     = tf.Variable(tf.random_uniform([20,4], minval=0, maxval=1, dtype=tf.float32, seed=1234), name = 'var')
boxes = tf.Variable(tf.random_uniform([20,4], minval=0, maxval=128, dtype=tf.float32, seed=1234), name = 'var')
fill8 = tf.zeros([8,4], dtype = tf.float32)
init  = tf.global_variables_initializer().run()
```

```python
rng = tf.to_float(tf.range(num_detections))
rng = tf.expand_dims(rng, axis =1)
print(' Range: ',rng.shape)
print(rng.eval())
```

```python
print(fill8.eval())
print(' Pre concatenation a.shape', tf.shape(a).eval())
print(a.eval())
a = tf.concat([a, fill8],axis =0)
print(' Post concatenation a.shape', tf.shape(a).eval())
print(a.eval())

boxes = tf.concat([boxes, fill8], axis = 0)
print(' boxxes.shape', tf.shape(boxes).eval())
print(boxes.eval())
```

```python
np.set_printoptions(linewidth=100)
clas = tf.to_float(tf.argmax(a,axis=1))
clas = tf.expand_dims(clas,axis=1)
print(' classes with highest scores:',clas.shape)
print(clas.eval())
pred = tf.reduce_max(a,axis=1, keep_dims=True)   # (32,)
print(' highest scores:',pred.shape)
# print('amax.shape ', amax.shape)
a_boxes = tf.concat([rng, pred, boxes, clas], axis=1)
print(' resulting tensor : a_boxes', a_boxes.shape)
print(a_boxes.eval())
```

###  Sort Tensor by class id
This is column 6

```python
sort_vals, sort_indcs = tf.nn.top_k(-a_boxes[:,6], k=a_boxes.shape[0])
print(sort_vals.eval())
print(sort_indcs.eval())
print(tf.shape(a_boxes).eval(), tf.shape(sort_indcs).eval())

srtd_boxes = tf.gather(a_boxes, sort_indcs)
print(' srtd_boxes shaoe: ', tf.shape(srtd_boxes).eval())
print(srtd_boxes.eval())

## - do we really need this ? srtd_boxes[:,6] gives us the class ptrs needed 
##   for dynamic_partitioning 

uniq, cls_ptrs = tf.unique(srtd_boxes[:,6])
print(uniq.shape, cls_ptrs.shape)
print(' Unqiue class ids: ', uniq.eval())
print(' Row / class mapping: ', cls_ptrs.eval())
```

### Dynamic Partitioning 

```python
cls_ptrs = tf.to_int32(srtd_boxes[:,6])
print('shape of srtd_boxes', srtd_boxes.shape)
print('cls_ptrs         ', cls_ptrs.eval())

splitted = tf.dynamic_partition(srtd_boxes, cls_ptrs, num_classes)

print(type(splitted), ' Partitioned to ', len(splitted), 'tensors')

for spl in splitted:
    print(' ------- ')
    print(spl.eval())
```

###  Sort again - this time by score within each class 

```python
# for idx in range(len(scatter)):
for idx, ten in enumerate(splitted):
#     ten = tf.squeeze(ten)
    ten_reordered = tf.gather(ten, tf.nn.top_k(ten[:,1], k=tf.shape(ten)[0]).indices)

    #     ten = scatter[idx]
    print(' item ', idx,' shape ',ten_reordered.shape, 'shape:', tf.shape(ten_reordered).eval(),
          'rank: ', tf.rank(ten_reordered).eval())
#     print(ten_reordered.eval())
#     padding = num_detections - tf.shape(ten_reordered)[0]
#     print(' num of padding rows tt add ', padding.eval())
#     print(ten.eval())  
#     ten = tf.pad(ten_reordered,[[0,padding],[0,0]]) 
#     print(' shape after padding : ', tf.shape(ten).eval())
    print(ten_reordered.eval())
    splitted[idx] = ten_reordered
    
```

```python
for idx, ten in enumerate(splitted):
    print(' item ', idx,' shape ',ten.shape, 'shape:', tf.shape(ten).eval(), 'rank: ', tf.rank(ten).eval())
    print(ten.eval())
 
```

###  Pad each individual tensor to same length and stack

After this we will have a a list of tensors consisting of NUM_CLASSES  tensors each with a shape of (num_detections x 7) 

```python
for idx, ten in enumerate(splitted):
    print(' item ', idx,' shape ',ten.shape, 'shape:', tf.shape(ten).eval(), 'rank: ', tf.rank(ten).eval())
    padding = num_detections - tf.shape(ten)[0]
    print(' num of padding rows tt add ', padding.eval())
#   print(ten.eval())  
    ten = tf.pad(ten,[[0,padding],[0,0]]) 
    print(' shape after padding : ', tf.shape(ten).eval())
    print(ten.eval())
    splitted[idx] = ten
    
```

###  Finally - stack the list of tensors into one tensor

Shape of this tensor is NUM_CLASSES x NUM_MAX_DETECTIONS x NUM_COLUMNS (7)

```python
stacked = tf.stack(splitted, axis = 0)
print('shape is ', tf.shape(stacked).eval())
print(stacked.eval())
```
